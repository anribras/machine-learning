### 将图形嵌入式到notebook中

```python
%matplotlib inline
%config InlineBackend.figure_format='retina'
import matplotlib.pyplot as plt;
import numpy as np;

x=np.linspace(0,1,300);
plt.figure(1);
for w in range(2,6,2):
    plt.plot(x,np.sin(np.pi*x)*np.sin(np.pi*2*w));
x = np.linspace(0, 1, 1000)**1.5;
plt.figure(2);
plt.hist(x);
```
